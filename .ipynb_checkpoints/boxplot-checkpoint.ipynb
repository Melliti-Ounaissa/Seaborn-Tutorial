```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.Series([5, 7, 8, 10, 12, 14, 16])
df = pd.DataFrame(data, columns=["data"])
df
```

```python
df.describe()
```

```python
sns.set_theme(style='darkgrid')
plt.figure(figsize=(4,5))
sns.boxplot(y='data', data=df, showmeans=True, meanprops={"marker":"o", "markerfacecolor":"white", "markeredgecolor":"black"})
```

The boxplot is used to represent the 3 quartiles (25%, 50% median, 75%) as a box.

The caps indicate the minumum/maximum value.

As for the small circle, it's the mean.
