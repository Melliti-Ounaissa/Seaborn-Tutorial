## Violinplot

It is a mix between the mirrored and vertically rotate KDE plot and the boxplot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.Series([5, 7, 8, 10, 12, 14, 16])
df = pd.DataFrame(data, columns=["data"])
df
```

```python
plt.figure(figsize=(16,5))
sns.violinplot(x='data', data=df)
```
